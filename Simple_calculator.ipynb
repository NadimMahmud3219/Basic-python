```python
print("what do you want to do??\n for addition select 1 \n for subtraction select 2 \n for multiplication select 3 \n for divition enter 4")
operation = int(input())

num1 = int(input("enter first number"))
num2 = int(input("enter second number"))

if operation == 1:
    print(" Addition = ",num1+num2)
elif operation == 2:
    print(" Subtraction = ",num1-num2)
elif operation == 3:
    print(" Multiplication = ",num1*num2)
elif operation == 4:
    if num2 == 0:
        print("enter number grater than zero!!")
    else:
        print(" Division = ",num1/num2)
else:
    print("invalid input")
```
